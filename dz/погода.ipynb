```python
day_weather_10_years = [17, 8, 4, 24, 9, 12, 13, 15, 19, 14]
forecast = 12

average_temp = sum(day_weather_10_years) / len(day_weather_10_years)

if average_temp > forecast:
    compare = "выше"
elif average_temp < forecast:
    compare = "ниже"
else:
    compare = "равна"

print(f"Средняя температура за 10 лет: {average_temp:.1f}°C. "
      f"Прогноз на 25 сентября: {forecast}°C. "
      f"Температура {compare} прогноза.")
```
